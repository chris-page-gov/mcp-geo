# Map Delivery Option Tracker

Loads the Playwright trial observations and summarizes map delivery coverage by trial and browser.

```python
from pathlib import Path
import json

root = Path('..').resolve().parents[0]
obs_path = root / 'evidence' / 'logs' / 'playwright_trials_observations.jsonl'
records = [json.loads(line) for line in obs_path.read_text().splitlines() if line.strip()]
print(f'Loaded {len(records)} observation rows from {obs_path}')
```

```python
latest = {}
for row in records:
    key = (row['trialId'], row['browser'])
    if key not in latest or row['timestamp'] > latest[key]['timestamp']:
        latest[key] = row

for trial in sorted({k[0] for k in latest}):
    print(f'\n{trial}')
    for browser in sorted(k[1] for k in latest if k[0] == trial):
        row = latest[(trial, browser)]
        print(f'  - {browser}: {row["status"]}')
```
